# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. It has built in themes for styling Matplotlib graphics

Note: Seaborn is meant to serve as a complement to matplotlib, and not a replacement.


```python
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
```

### The data consists of the following features :

- total_bill: The total bill paid by the customer.
- tip: The tip provided by the customer.
- sex: The gender of the customer.
- smoker: If the customer is a smoker or not.
- day: The day of the week when the observation was made.
- time: The time of the observation, whether at lunch or dinner etc.
- size: The size of the group whether there were multiple members.

```python
# Load in the data set
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
# Print out the summary statistics for the quantitative variables
df.describe()
```

### One Variable

```python
# Create a boxplot of the tips amounts
sns.boxplot(df["tip"]).set_title("Box plot of the Tip")
plt.show()
```

```python
# Plot a histogram of the total bill
ax = sns.distplot(df["total_bill"],hist=True ,kde=False)
ax.set(ylabel='Probability', title='Histogram')
plt.show()
```

```python
# Plot a histogram of the total bill
ax = sns.distplot(df["total_bill"],hist=True ,kde=True)
ax.set(ylabel='Probability', title='Histogram')
plt.show()
```

### Two Variables

```python
# Plot a Scatter Plot of both the total bill and the tips'
sns.scatterplot(x="total_bill", y="tip", data=df)
```

```python
# Plot a histogram of both the total bill and the tips'
sns.distplot(df["total_bill"], kde = False)
sns.distplot(df["tip"], kde = False).set_title("Histogram of Both Tip Size and Total Bill")
plt.show()
```

```python
# boxplot of the tips grouped by smoking status
sns.boxplot(x = df["tip"], y = df["smoker"])
plt.show()
```

```python

# categorical plot of the tips grouped by smoking status
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=df);
```

```python
# boxplot of the tips grouped by time of day
sns.boxplot(x = df["tip"], y = df["time"])
```

```python
# Create a BarPlot of the tips grouped by time of day
ax = sns.barplot(x="time", y="tip", data=df,
                 order=["Dinner", "Lunch"],ci=None)
plt.show()
```

FacetGrid object takes a dataframe as input and the names of the variables 
that will form the row, column, or hue dimensions of the grid. The variables should be 
categorical and the data at each level of the variable will be used for a facet along that axis.

```python
g = sns.FacetGrid(df, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
#Create a BarPlot Total_bill and Day
sns.barplot(x="day", y="total_bill", data=df, ci=None)
```

### 3 Variables 

```python
g = sns.FacetGrid(df, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=df ,ci=None)
plt.show()
```

## More than 3 Variables

```python
#creating The Bar plot 
g = sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=df, kind="bar",
                height=4, aspect=.7, ci=None);
```
