```python
def countVowels(s) :
    count = 0
    for i in s.lower() :
        if i in 'aeiou' :
            count += 1
    return count

print(countVowels('happy'))
print(countVowels('excited'))
print(countVowels('spectacular'))
```

```python
df.drop(['job', 'month', 'pdays', 'previous', 'poutcome'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df_unique = {col : df[col].nunique() for col in df.columns}
df_unique
```

```python
l = [i for i, j in df_unique.items() if j == 2]
l
```

```python
df[l] = df[l].replace({'no' : 0, 'yes' : 1}).astype('category')
df
```

```python
df.dtypes
```

```python
cat = df.select_dtypes(include = 'object').astype('category').columns
df[cat]
```

```python
dummy = pd.get_dummies(df[cat])
dummy
```

```python
df.drop(cat, axis = 1, inplace = True)

df = pd.concat([df, dummy], axis = 1)
```

```python
df.dtypes
```

```python
num = df.select_dtypes(include = 'int64').columns
num
```

```python
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()

std = pd.DataFrame(standardizer.fit_transform(df[num]), columns = num)
std
```

```python
df.drop(std, axis = 1, inplace = True)
pd.concat([std, df], axis = 1)
```
