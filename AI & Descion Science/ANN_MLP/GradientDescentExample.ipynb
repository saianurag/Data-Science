
1.  Initial random weight ie. w1 w2

2.  predict with random weight as defined in (1)
    ypred = w1x1 +w2x2
    
3.  Calculate error  based on metrics i.e. mse
    error(E) = (y-ypred)**2  = (y- (w1x1 +w2x2))**2
    
4.  if optimiser is sgd i will take
    dEw1  = 2*(y-ypred)*(-x1)
    dEw2  = 2*(y-ypred)*(-x2)
    
5. update the weight based of dEw and learning rate(lr - 0.001)
    w1 = w1 - lr*dEw1
    w2 = w2 - lr*dEw2
    
6. got to point no 2

### Gradient Descent with learning rate and back propagation


```python
lr = 0.01
n = 20
w1,w2 = 0.3, 0.3
for epoch in range(n):
    ypred = w1*x1 +w2*x2
    err = (y-ypred)**2
    print("epoch ",epoch,": w1: ", w1, "w2: ",w2, "error: ",err)
    #dEw1 = 2 *(y-ypred)
    
    dEw1 = 2*(y-ypred)*(-x1)
    dEw2 = 2*(y-ypred)*(-x2)
    w1 = w1 - lr*dEw1
    w2 = w2 - lr*dEw2
```

### Gradient Descent with learning rate and early stopping

```python
x1, x2 = 2,5
y = 31
w1,w2 = 0.3, 0.3
lr = 0.01
n = 20
for epoch in range(n):
    ypred = w1*x1 +w2*x2
    err = (y-ypred)**2
    print("epoch ",epoch,": ",err)
    #dEw1 = 2 *(y-ypred)
    
    dEw1 = 2*(y-ypred)*(-x1)
    dEw2 = 2*(y-ypred)*(-x2)
    w1 = w1 - lr*dEw1
    w2 = w2 - lr*dEw2
    
    if err < 0.0005:
        break
```

```python
n = 200 # Total epoch

df = pd.DataFrame({"x1":[4,9], "x2":[3,3], "y":[11,21]})
x = df.iloc[:, 0:2].values
y = df.iloc[:, 2].values
w = np.array([5,5])
lr = 0.001
for epoch in range(n):
    ypred = np.dot(x,w)
    err = 1/2 * np.sum((y -ypred)**2)
    print("epoch ",epoch,": w1: ", w[0], "w2: ",w[1], "error: ",err)
    #print("Prediction ", ypred)
    dEw1 = 2*np.dot(((y-ypred)), -x[:,0])
    dEw2 = 2*np.dot(((y-ypred)), -x[:,1])
    w1 = w[0] - lr*dEw1
    w2 = w[1] - lr*dEw2
   
    w = np.array([w1, w2])
    
    if err < 0.001:
        break
    #print(w)
```
